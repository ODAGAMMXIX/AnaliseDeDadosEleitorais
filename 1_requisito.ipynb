```python
import pandas as pd
import plotly.graph_objects as go
from cobertura import abrangencia

csvEleitorSP = './dados/perfil_eleitor_secao_ATUAL_SP.csv'

count = 0
query = ""

for abr_code in abrangencia:
    query += 'CD_MUNICIPIO == ' + str(abr_code["codigo"])
    count += 1
    if count < len(abrangencia):
        query += 'or '

data = pd.read_csv(csvEleitorSP, sep=";", encoding="ISO-8859-1",skiprows=0)
data.query(query, inplace = True)
```

```python
escolaridade = data.groupby(['DS_GRAU_ESCOLARIDADE'])['DS_GRAU_ESCOLARIDADE'].count().plot(kind='pie')
print(data.groupby(['DS_GRAU_ESCOLARIDADE'])['DS_GRAU_ESCOLARIDADE'].count())
```

```python
faixa_etaria = data.groupby(['DS_FAIXA_ETARIA'])['DS_FAIXA_ETARIA'].count().plot(kind='barh')
print(data.groupby(['DS_FAIXA_ETARIA'])['DS_FAIXA_ETARIA'].count())
```

```python
estado_civil = data.groupby(['DS_ESTADO_CIVIL'])['DS_ESTADO_CIVIL'].count().plot(kind='barh')
print(data.groupby(['DS_ESTADO_CIVIL'])['DS_ESTADO_CIVIL'].count())
```
